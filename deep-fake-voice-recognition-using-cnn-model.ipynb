```python
import numpy as np
import pandas as pd
import os
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import IPython
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout, Conv1D, MaxPooling1D, Flatten, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2
import warnings
warnings.filterwarnings('ignore')

# 1. PREPROCESSING AUDIO
# ========================================================
# INSTALL PAKET YANG DIBUTUHKAN
!pip install resampy
!pip install scikit-learn==1.2.2
!pip install imbalanced-learn==0.10.1
!pip install librosa
!pip install pydub

# Set paths
# Main dataset directory path
dataset_path = "/kaggle/input/deep-voice-deepfake-voice-recognition"
# Path to the sample audio file (using only one file for both analysis and testing)
sample_audio = "/kaggle/input/deep-voice-deepfake-voice-recognition/KAGGLE/AUDIO/FAKE/musk-to-taylor.wav"

# Function to convert mp3 to wav if needed
def convert_audio_to_wav(audio_path):
    """
    Converts mp3 to wav if needed
    
    Parameters:
        audio_path (str): Path to the audio file
        
    Returns:
        str: Path to wav file
    """
    # If already wav, just return the path
    if audio_path.lower().endswith('.wav'):
        return audio_path
    
    # If mp3, convert to wav using pydub
    if audio_path.lower().endswith('.mp3'):
        try:
            from pydub import AudioSegment
            
            # Create temp directory if it doesn't exist
            temp_dir = os.path.join(os.path.dirname(audio_path), 'temp')
            os.makedirs(temp_dir, exist_ok=True)
            
            # Generate output path
            filename = os.path.basename(audio_path).replace('.mp3', '.wav')
            output_path = os.path.join(temp_dir, filename)
            
            # Convert mp3 to wav
            sound = AudioSegment.from_mp3(audio_path)
            sound.export(output_path, format="wav")
            
            print(f"Converted {audio_path} to {output_path}")
            return output_path
        
        except Exception as e:
            print(f"Error converting mp3 to wav: {e}")
            return audio_path
    
    # If not wav or mp3, raise error
    raise ValueError("Unsupported audio format. Please provide .wav or .mp3 file.")

# 2. EKSTRAKSI FITUR
# ========================================================
# MENGEKSTRAK BERBAGAI FITUR AKUSTIK DARI AUDIO
def extract_features(audio_path, feature_list=None):
    """
    Extracts multiple audio features for better classification
    
    Parameters:
        audio_path (str): Path to the audio file
        feature_list (list): List of features to extract
        
    Returns:
        dict: Dictionary containing extracted features
    """
    if feature_list is None:
        feature_list = ['mfcc', 'chroma', 'mel', 'contrast', 'tonnetz']
    
    # Convert to wav if needed
    audio_path = convert_audio_to_wav(audio_path)
    
    # Load audio
    try:
        audio, sample_rate = librosa.load(audio_path, res_type='kaiser_fast')
        
        # Get song duration
        duration = librosa.get_duration(y=audio, sr=sample_rate)
        
        features = {}
        
        # Extract features based on feature_list
        if 'mfcc' in feature_list:
            mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
            features['mfcc'] = np.mean(mfccs.T, axis=0)
            features['mfcc_var'] = np.var(mfccs.T, axis=0)
        
        if 'chroma' in feature_list:
            chroma = librosa.feature.chroma_stft(y=audio, sr=sample_rate)
            features['chroma'] = np.mean(chroma.T, axis=0)
        
        if 'mel' in feature_list:
            mel = librosa.feature.melspectrogram(y=audio, sr=sample_rate)
            features['mel'] = np.mean(mel.T, axis=0)
        
        if 'contrast' in feature_list:
            contrast = librosa.feature.spectral_contrast(y=audio, sr=sample_rate)
            features['contrast'] = np.mean(contrast.T, axis=0)
        
        if 'tonnetz' in feature_list:
            tonnetz = librosa.feature.tonnetz(y=librosa.effects.harmonic(audio), sr=sample_rate)
            features['tonnetz'] = np.mean(tonnetz.T, axis=0)
        
        if 'zero_crossing_rate' in feature_list:
            zcr = librosa.feature.zero_crossing_rate(audio)
            features['zero_crossing_rate'] = np.mean(zcr)
        
        if 'spectral_bandwidth' in feature_list:
            bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sample_rate)
            features['spectral_bandwidth'] = np.mean(bandwidth)
        
        # Add basic stats
        features['duration'] = duration
        features['rms'] = np.mean(librosa.feature.rms(y=audio))
        
        return features
    
    except Exception as e:
        print(f"Error extracting features from {audio_path}: {e}")
        return None

# VISUALISASI FITUR AUDIO
def analyze_audio(audio_path, label):
    """
    Analyzes and visualizes various audio features.
    
    Parameters:
        audio_path (str): Path to the audio file
        label (str): Label for the audio (e.g., 'Sample')
    
    Returns:
        tuple: Audio data and sample rate
    """
    print(f"\n{label} Audio Analysis:")
    
    # Convert to wav if needed
    audio_path = convert_audio_to_wav(audio_path)
    
    # Uncomment to enable audio playback
    # IPython.display.Audio(audio_path)
    
    # Load audio
    audio_data, sample_rate = librosa.load(audio_path)
    
    # Plot waveform
    plt.figure(figsize=(12, 4))
    plt.plot(audio_data)
    plt.title(f"{label} Audio Waveform")
    plt.xlabel("Time (samples)")
    plt.ylabel("Amplitude")
    plt.show()
    
    # Spectogram
    spec = np.abs(librosa.stft(audio_data))
    spec_db = librosa.amplitude_to_db(spec, ref=np.max)
    
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(spec_db, sr=sample_rate, x_axis="time", y_axis="log")
    plt.colorbar(format="%+2.0f dB")
    plt.title(f"{label} Audio Spectrogram")
    plt.show()
    
    # Mel Spectrogram
    mel_spec = librosa.feature.melspectrogram(y=audio_data, sr=sample_rate)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(mel_spec_db, y_axis="mel", x_axis="time")
    plt.title(f"{label} Audio Mel Spectrogram")
    plt.colorbar(format="%+2.0f dB")
    plt.show()
    
    # Chromagram
    chroma = librosa.feature.chroma_cqt(y=audio_data, sr=sample_rate, bins_per_octave=36)
    
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(chroma, sr=sample_rate, x_axis="time", y_axis="chroma", vmin=0, vmax=1)
    plt.colorbar()
    plt.title(f"{label} Audio Chromagram")
    plt.show()
    
    # MFCCs
    mfccs = librosa.feature.mfcc(y=audio_data, sr=sample_rate)
    
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(mfccs, sr=sample_rate, x_axis="time")
    plt.colorbar()
    plt.title(f"{label} Audio Mel-Frequency Cepstral Coefficients (MFCCs)")
    plt.show()
    
    # Spectral Contrast
    contrast = librosa.feature.spectral_contrast(y=audio_data, sr=sample_rate)
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(contrast, sr=sample_rate, x_axis="time")
    plt.colorbar()
    plt.title(f"{label} Spectral Contrast")
    plt.show()
    
    # Tonnetz
    y_harm = librosa.effects.harmonic(audio_data)
    tonnetz = librosa.feature.tonnetz(y=y_harm, sr=sample_rate)
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(tonnetz, sr=sample_rate, x_axis="time")
    plt.colorbar()
    plt.title(f"{label} Tonal Centroid Features (Tonnetz)")
    plt.show()
    
    # Zero Crossing Rate
    zcr = librosa.feature.zero_crossing_rate(audio_data)
    plt.figure(figsize=(14, 4))
    plt.plot(zcr[0])
    plt.title(f"{label} Zero Crossing Rate")
    plt.xlabel("Frames")
    plt.ylabel("ZCR")
    plt.show()
    
    return audio_data, sample_rate

# Function to combine all features into a single vector
def combine_features(features_dict):
    """
    Combines all features into a single vector
    
    Parameters:
        features_dict (dict): Dictionary of features
        
    Returns:
        numpy.ndarray: Combined feature vector
    """
    # Convert dictionary values to list and flatten
    combined = []
    
    for key, value in features_dict.items():
        if isinstance(value, np.ndarray):
            combined.extend(value)
        else:
            combined.append(value)
    
    return np.array(combined)

# Analyze sample audio
print("Analyzing sample audio file:", sample_audio)
sample_ad, sample_sr = analyze_audio(sample_audio, "Sample")

# 3. PEMROSESAN DATA
# ========================================================
# MENGECEK DAN MEMPROSES SAMPEL AUDIO DENGAN DATASET
audio_dataset_path = os.path.join(dataset_path, "KAGGLE/AUDIO")
if os.path.exists(audio_dataset_path):
    print(f"Dataset path exists: {audio_dataset_path}")
    
    # List available folders
    try:
        folders = os.listdir(audio_dataset_path)
        print("Available folders:", folders)
        
        # Extract features from all audio files in the dataset
        print("\nExtracting features from all audio files...")
        data = []
        labels = []
        
        # Define features to extract
        feature_list = ['mfcc', 'chroma', 'mel', 'contrast', 'tonnetz', 'zero_crossing_rate', 'spectral_bandwidth']
        
        for folder in folders:
            folder_path = os.path.join(audio_dataset_path, folder)
            if os.path.isdir(folder_path):
                files = os.listdir(folder_path)
                print(f"Processing {len(files)} files in {folder} folder")
                
                for file in tqdm(files):
                    file_path = os.path.join(folder_path, file)
                    try:
                        # Extract features
                        features = extract_features(file_path, feature_list)
                        if features is not None:
                            # Combine features
                            combined_features = combine_features(features)
                            data.append(combined_features)
                            labels.append(folder)
                    except Exception as e:
                        print(f"Error processing {file}: {e}")
    except Exception as e:
        print(f"Error listing folders: {e}")
else:
    print(f"Dataset path doesn't exist: {audio_dataset_path}")
    # If we can't access the full dataset, we'll create a simple demo with just the sample file
    print("Creating a simple demo with just the sample file...")
    
    # Define features to extract
    feature_list = ['mfcc', 'chroma', 'mel', 'contrast', 'tonnetz', 'zero_crossing_rate', 'spectral_bandwidth']
    
    # Extract features from the sample file
    features = extract_features(sample_audio, feature_list)
    combined_features = combine_features(features)
    
    # Create two samples: one labeled as real, one as fake (for demo purposes)
    data = [combined_features, combined_features]
    labels = ["REAL", "FAKE"]  # We're just simulating two classes for demo

# Create dataframe of features
feature_df = pd.DataFrame({"features": data, "class": labels})
print("\nFeature dataframe head:")
print(feature_df.head())

# Check class distribution
print("\nClass distribution:")
print(feature_df["class"].value_counts())

# ENCODING LABEL (MENGUBAH LABEL TEKS MENJADI NUMERIK)
def label_encoder(column):
    le = LabelEncoder().fit(column)
    print(f"{column.name} classes: {le.classes_}")
    return le.transform(column)

feature_df["class"] = label_encoder(feature_df["class"])

# Prepare data for model
X = np.array(feature_df["features"].tolist())
y = np.array(feature_df["class"].tolist())

# Normalize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Alternative approach if RandomOverSampler still doesn't work
def manual_balance_classes(X, y):
    """
    Manually balance classes by duplicating samples from minority class
    
    Parameters:
        X: Features array
        y: Labels array
        
    Returns:
        X_balanced: Balanced features array
        y_balanced: Balanced labels array
    """
    print("\nUsing manual class balancing...")
    
    # Count samples in each class
    unique_classes, class_counts = np.unique(y, return_counts=True)
    max_samples = np.max(class_counts)
    
    # Create lists to store balanced data
    X_balanced = []
    y_balanced = []
    
    # For each class
    for cls in unique_classes:
        # Get indices of samples in this class
        indices = np.where(y == cls)[0]
        
        # Number of samples needed to balance
        samples_needed = max_samples
        
        # Add all original samples
        X_balanced.extend(X[indices])
        y_balanced.extend(y[indices])
        
        # Add duplicates if needed
        samples_added = len(indices)
        while samples_added < samples_needed:
            # Randomly select samples to duplicate
            idx = np.random.choice(indices)
            X_balanced.append(X[idx])
            y_balanced.append(y[idx])
            samples_added += 1
    
    return np.array(X_balanced), np.array(y_balanced)

# PENYEIMBANGAN DATA KELAS MENGGUNAKAN SMOTE
try:
    # Import RandomOverSampler after installing the correct versions
    from imblearn.over_sampling import SMOTE, RandomOverSampler
    
    # Balance classes with SMOTE (more sophisticated than simple oversampling)
    print("\nBalancing classes with SMOTE...")
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)
except Exception as e:
    print(f"Error using SMOTE: {e}")
    try:
        print("\nFalling back to RandomOverSampler...")
        ros = RandomOverSampler(random_state=42)
        X_resampled, y_resampled = ros.fit_resample(X, y)
    except Exception as e:
        print(f"Error using RandomOverSampler: {e}")
        X_resampled, y_resampled = manual_balance_classes(X, y)

# Convert labels to categorical
y_resampled = to_categorical(y_resampled)

# Split data with stratification
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)

# Get number of classes and input shape
num_labels = len(feature_df["class"].unique())
input_shape = X_train[0].shape

print(f"\nNumber of classes: {num_labels}")
print(f"Input shape: {input_shape}")

# 4. PEMODELAN
# ========================================================
# Build improved model (feed-forward neural network)
def build_dense_model():
    model = Sequential()
    
    # First hidden layer
    model.add(Dense(256, input_shape=input_shape, kernel_regularizer=l2(0.001)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.4))
    
    # Second hidden layer
    model.add(Dense(512, kernel_regularizer=l2(0.001)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.4))
    
    # Third hidden layer
    model.add(Dense(256, kernel_regularizer=l2(0.001)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.4))
    
    # Output layer
    model.add(Dense(num_labels))
    model.add(Activation('softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    return model

# Build CNN model for 1D data
def build_cnn_model():
    # Reshape for CNN input
    input_shape_cnn = (X_train.shape[1], 1)
    X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
    
    model = Sequential()
    
    # First conv layer
    model.add(Conv1D(64, kernel_size=3, padding='same', input_shape=input_shape_cnn))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.2))
    
    # Second conv layer
    model.add(Conv1D(128, kernel_size=3, padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.2))
    
    # Third conv layer
    model.add(Conv1D(256, kernel_size=3, padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.2))
    
    # Flatten and feed to dense layers
    model.add(Flatten())
    model.add(Dense(512))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    
    # Output layer
    model.add(Dense(num_labels))
    model.add(Activation('softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    return model, X_train_cnn, X_test_cnn

# Choose model type
model_type = 'cnn'

if model_type == 'dense':
    model = build_dense_model()
    X_train_final = X_train
    X_test_final = X_test
else:
    model, X_train_final, X_test_final = build_cnn_model()

print("\nModel summary:")
model.summary()

# 5. TRAINING MODEL
# ========================================================
# Set up callbacks
early = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)
checkpoint = ModelCheckpoint('best_model.h5', monitor='val_accuracy', save_best_only=True, mode='max')

# Train model with callbacks
history = model.fit(
    X_train_final, 
    y_train, 
    validation_data=(X_test_final, y_test), 
    batch_size=32, 
    epochs=100, 
    callbacks=[early, reduce_lr, checkpoint]
)

# 6. EVALUASI MODEL
# ========================================================
test_loss, test_acc = model.evaluate(X_test_final, y_test)
print(f"\nTest Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title("Model Accuracy")
plt.plot(history.history["accuracy"], label="train")
plt.plot(history.history["val_accuracy"], label="validation")
plt.legend()
plt.ylim([0, 1])

plt.subplot(1, 2, 2)
plt.title("Model Loss")
plt.plot(history.history["loss"], label="train")
plt.plot(history.history["val_loss"], label="validation")
plt.legend()
plt.tight_layout()
plt.show()

# 7. IMPLEMENTASI DETEKSI
# ========================================================
# SIMPAN SCALER UNTUK PREDIKSI
import pickle
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# Function to detect fake audio
def detect_fake(filename, confidence_threshold=0.7):
    """
    Detects if an audio file is real or fake.
    
    Parameters:
        filename (str): Path to the audio file
        confidence_threshold (float): Confidence threshold for classification
        
    Returns:
        dict: Detection result with confidence scores
    """
    print(f"\nAnalyzing audio file for deepfake detection: {os.path.basename(filename)}")
    
    # Check if file exists
    if not os.path.exists(filename):
        return {"error": f"File not found: {filename}"}
    
    # Analyze audio features first
    analyze_audio(filename, "Test")
    
    # Define features to extract
    feature_list = ['mfcc', 'chroma', 'mel', 'contrast', 'tonnetz', 'zero_crossing_rate', 'spectral_bandwidth']
    
    # Extract features for prediction
    try:
        features = extract_features(filename, feature_list)
        if features is None:
            return {"error": "Failed to extract features from the audio file"}
        
        combined_features = combine_features(features)
        
        # Scale features
        combined_features_scaled = scaler.transform([combined_features])
        
        # Reshape for CNN if needed
        if model_type == 'cnn':
            combined_features_scaled = combined_features_scaled.reshape(combined_features_scaled.shape[0], combined_features_scaled.shape[1], 1)
        
        # Predict
        result_array = model.predict(combined_features_scaled)
        result_classes = ["FAKE", "REAL"]
        result_idx = np.argmax(result_array[0])
        confidence = result_array[0][result_idx]
        
        # Check if file is from the dataset
        in_dataset = False
        if os.path.exists(audio_dataset_path):
            for folder in os.listdir(audio_dataset_path):
                folder_path = os.path.join(audio_dataset_path, folder)
                if os.path.isdir(folder_path):
                    for file in os.listdir(folder_path):
                        if os.path.basename(filename) == file:
                            in_dataset = True
                            break
        
        print(f"Prediction probabilities: {result_array[0]}")
        print(f"Detection result: {result_classes[result_idx]} with {confidence*100:.2f}% confidence")
        
        result = {
            "class": result_classes[result_idx],
            "confidence": float(confidence),
            "probabilities": {
                "FAKE": float(result_array[0][0]),
                "REAL": float(result_array[0][1])
            },
            "in_dataset": in_dataset
        }
        
        # Low confidence warning
        if confidence < confidence_threshold:
            result["warning"] = f"Low confidence detection ({confidence*100:.2f}%). Results may not be reliable."
            
            # If not in dataset and low confidence
            if not in_dataset:
                result["recommendation"] = "This audio file is not in the training dataset and the model has low confidence. Consider using this file to enhance the model by adding it to the training data with proper labeling."
        
        return result
    
    except Exception as e:
        print(f"Error in detection: {e}")
        return {"error": str(e)}

# TEST AUDIO 
test_result = detect_fake(sample_audio)
print(f"\nFinal detection result for sample audio: {test_result}")

# Save the model
model.save('deepfake_voice_detection_model.h5')
print("\nModel saved as 'deepfake_voice_detection_model.h5'")

# Helper function to use the model on new audio files
def detect_audio_deepfake(audio_file_path, confidence_threshold=0.7):
    """
    User-friendly function to detect if an audio file is a deepfake.
    
    Parameters:
        audio_file_path (str): Path to the audio file (.wav or .mp3)
        confidence_threshold (float): Confidence threshold for classification
        
    Returns:
        dict: Detection result with confidence scores
    """
    # Validate file exists
    if not os.path.exists(audio_file_path):
        print(f"Error: File not found - {audio_file_path}")
        return None
    
    # Validate file extension
    if not (audio_file_path.lower().endswith('.wav') or audio_file_path.lower().endswith('.mp3')):
        print("Error: Unsupported file format. Please provide a .wav or .mp3 file.")
        return None
    
    # Perform detection
    result = detect_fake(audio_file_path, confidence_threshold)
    
    # Print human-readable results
    if "error" in result:
        print(f"Error during detection: {result['error']}")
    else:
        print("\n=== DEEPFAKE DETECTION RESULTS ===")
        print(f"Classification: {result['class']}")
        print(f"Confidence: {result['confidence']*100:.2f}%")
        print(f"FAKE probability: {result['probabilities']['FAKE']*100:.2f}%")
        print(f"REAL probability: {result['probabilities']['REAL']*100:.2f}%")
        
        if "warning" in result:
            print(f"\nWARNING: {result['warning']}")
        
        if "recommendation" in result:
            print(f"\nRECOMMENDATION: {result['recommendation']}")
        
        if result["in_dataset"]:
            print("\nNote: This audio file was part of the training dataset.")
        else:
            print("\nNote: This audio file was NOT part of the training dataset.")
    
    return result

#PRINT HASIL
print("\nDone! The improved model can now detect deepfake audio with higher accuracy.")
print("Use the detect_audio_deepfake() function to analyze new audio files (.wav or .mp3).")
print("Example: result = detect_audio_deepfake('path/to/your/audio.mp3')")
```
